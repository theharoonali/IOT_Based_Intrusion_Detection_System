```python

# Split the data into training and testing sets
X_train, X_test, y1_train, y1_test, y2_train, y2_test = train_test_split(X, y1, y2, test_size=0.01)

```

```python
def isqrt(n):
    x = n
    y = (x + 1) // 2
    while y < x:
        x = y
        y = (x + n // x) // 2
    return x
```

```python
k = isqrt(len(X))
```

```python

# Create an instance of the KNeighborsClassifier class with n_neighbors=3
knn = KNeighborsClassifier(n_neighbors=k)
```

```python
# Wrap the kNN model in a MultiOutputClassifier to handle multiple output variables
multi_knn = MultiOutputClassifier(knn, n_jobs=-1)

# Fit the kNN model to the training data
multi_knn.fit(X_train, np.column_stack((y1_train, y2_train)))
```
